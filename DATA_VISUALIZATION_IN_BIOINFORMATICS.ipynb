**Step2: Load and View the CSV File**

```python
import pandas as pd

data = pd.read_csv("gene_data.csv")
print("ðŸ“„ Preview of Your Data:")
data.head()
```

**Step 3: Create Graphs (Bar Plot & GC Content Analysis)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Bar plot: Gene Length
plt.figure(figsize=(10, 5))
sns.barplot(x="Gene Name", y="Length (bp)", data=data, palette="viridis")
plt.title("Gene Length Comparison")
plt.xlabel("Gene")
plt.ylabel("Length (bp)")
plt.show()

# Line plot: GC Content
plt.figure(figsize=(10, 5))
sns.lineplot(x="Gene Name", y="GC Content (%)", data=data, marker="o", color="green")
plt.title("GC Content Across Genes")
plt.xlabel("Gene")
plt.ylabel("GC Content (%)")
plt.show()
```

**Calculate GC Content with Biopython (Optional)
If you want to calculate GC content from the sequence itself:**

```python
from Bio.SeqUtils import gc_fraction
from Bio.Seq import Seq

sequence = Seq("ATGCGTACGTGCTAGCTAGCGTAGC")  # Replace with your gene sequence
gc = gc_fraction(sequence) * 100
print(f"GC Content: {gc:.2f}%")
```

**Bar Graph â€“ Gene-wise GC Content
This shows how much GC content each gene has. It helps you compare across different genes.**

```python
import matplotlib.pyplot as plt

genes = ['BRCA1', 'TP53', 'EGFR', 'VEGFA', 'INS']
gc_values = [53.1, 48.7, 56.2, 45.3, 51.8]  # Example GC%

plt.bar(genes, gc_values, color='skyblue')
plt.title('GC Content in Selected Genes')
plt.xlabel('Gene')
plt.ylabel('GC Content (%)')
plt.ylim(40, 60)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Pie Chart Nucleotide Composition (A, T, G, C) for 1 Gene
This shows what percentage of each nucleotide is present in a single gene.**

```python
from collections import Counter

sequence = "ATGCGTAGCGTTAGCGCTAGCTAGCGA"  # Replace with real sequence
counts = Counter(sequence.upper())

labels = ['A', 'T', 'G', 'C']
sizes = [counts.get(base, 0) for base in labels]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightgreen', 'skyblue'])
plt.title('Nucleotide Composition')
plt.axis('equal')
plt.show()
```
