**STEP 2: Install & Import Libraries**

```python
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**STEP 3: Load & Clean Data**

```python
data = pd.read_csv("gene_expression.csv")

# Show first 5 rows
print(data.head())

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())

# Fill or remove missing values (if any)
data.fillna(0, inplace=True)
```

**STEP 4: Create Line Plot**

```python
# Set gene names as index (optional, for cleaner plot)
data.set_index("Gene", inplace=True)

# Transpose for line plot (genes vs samples)
data_T = data.transpose()

# Plot
plt.figure(figsize=(10,6))
for gene in data_T.columns:
    plt.plot(data_T.index, data_T[gene], label=gene)

plt.title("Gene Expression Over Samples")
plt.xlabel("Samples")
plt.ylabel("Expression Level")
plt.legend()
plt.grid(True)
plt.show()
```

**STEP 5: Create Heatmap**

```python
plt.figure(figsize=(8,6))
sns.heatmap(data, cmap="YlGnBu", annot=True)
plt.title("Gene Expression Heatmap")
plt.xlabel("Samples")
plt.ylabel("Genes")
plt.show()
```
